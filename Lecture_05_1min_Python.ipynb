# **4️⃣ Constants in Python**

**A Constant is a value that**

 **should not change during the program**.

**Constants store important fixed information.**

**In Python, constants are**

 **written in CAPITAL LETTERS**

**to tell programmers**:

 **“This value should stay the same**.”

# **Real Life Examples**

**DAYS_IN_WEEK**   =    7	(**Fixed days**)

**COUNTRY_NAME** =  "**Pakistan**"	(**Fixed country**)

**SCHOOL_NAME** = "**ABC School**"	(**Permanent name**)

**Your birth date**  **usually never changes**.

```python
DATE_OF_BIRTH = "10 March"
PI = 3.14
MAX_STUDENTS = 50

print(DATE_OF_BIRTH)
print(PI)
print(MAX_STUDENTS)
```
