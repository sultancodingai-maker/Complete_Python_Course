# **1️⃣ Identifiers**

**Identifier means a name in Python.**

# **We** **use** **names** **for**:

▶ **Variables**

▶ **Functions**

▶ **Classes**

# **Rules**:

✅  **Use letters, numbers, Underscore** (_)

✅ **Cannot start with number**

✅ **No spaces**

✅ **Cannot use Python keywords**

✅ **Python is Case-Sensitive**

✅ **Use meaningful names**


# **Real World Example**

**Imagine a classroom.**

**Every student has a unique name.**

**Teacher uses names to identify students.**

**Python also uses identifiers to identify variables and functions.**

# **✅ Valid Identifiers**

```python
user_name = "Ali"      # Underscore is allowed
age = 20               # Starts with a letter
total_marks = 450      # Letters + underscore are allowed
my_city = "Karachi"      # Identifier can start with underscore
```

# **❌ Invalid Identifiers**

```python
2age = 20              # Cannot start with a number

user-name = "Ali"      # Hyphen (-) is not allowed

my city = "Karachi"    # Spaces are not allowed

for = 10               # 'for' is a Python keyword
```
