 # 1️⃣ **Type Checking**

▶**Checks the data type of**

 **a value or variable.**

# Function: **type()**

# **✅ Example**

```python
num = 10
price = 99.5
name = "Ali"
status = True

print(type(num))
print(type(price))
print(type(name))
print(type(status))
```

# 2️⃣ **Type Conversion:**

▶ **Changes a value from**

 **one data type to another.**

# **Functions:**

**int()**

**float()**

**str()**

**bool()**

```python
age = "25"
price = 99
score = 100

print("String to Integer:", int(age))
print("Integer to Float:", float(price))
print("Integer to String:", str(score))
```
