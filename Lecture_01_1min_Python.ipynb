 # 🐍 **What is Python?**

➡ **Python is a high-level programming language.**

➡ **Python is easy to learn and read, almost like English.**


➡ **It is used in AI, websites, games, Automation and data science.**


# **But how does Python actually work internally?**
▶ **When you write Python code, the computer does NOT understand it directly.**

▶ **Computers only understand 0 and 1.**

▶ **First, the Python Interpreter reads your code line by line.**

▶ **Then it converts your code into bytecode (a middle language).**

▶ **After that, the Python Virtual Machine (PVM) runs the bytecode and gives output.**

# **Human Code → Python Interpreter → Bytecode → PVM → Computer Output**
