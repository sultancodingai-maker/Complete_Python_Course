# **3️⃣ Variables**

**A Variable is used to store data in Python.**

**Think of a variable like a container or a box**
**that stores information**.

**You can change things inside the box anytime**.

# **Example**

```python
name = "Sara"
print("Name:",name)

age = 40
print("Age:",age)

name = "ALi"
print("Update Name:",name)
```

# **Real World Example**

**Think about a water bottle**


**Today it contains**:

#**Juice**

**Tomorrow you can empty it and fill**:

# **Water**

**The bottle stays the same**,
**but the content changes**.

**Variables work exactly like this**.

**The variable name stays the same**,

**but its value can change anytime**.
