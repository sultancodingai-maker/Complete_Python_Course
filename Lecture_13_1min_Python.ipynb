# 2️⃣ **Converting Input into Numbers**


▶ **input() always returns**

 **text (string)**.

▶ **Use int() or float() when**

**working with numbers.**

# **✅ Integer Input**

▶ **Use int() when you**

 **need whole numbers.**

# **✅ Example**

```python
age = int(input("Enter Your Age"))
print("Your Age Is: ", age)
```

# **✅ Float Input**

▶ **Use float() when you**

 **need decimal numbers.**

```python
height = float(input("Enter your height: "))

print(height)
```

# Quick Recap

- **Text** → `input()`
- **Whole Number** → `int(input())`
- **Decimal Number** → `float(input())`
