# **2️⃣ Numbers in Python**

**Python mainly uses 3 types of numbers:**

**int → Whole numbers**

**float → Decimal numbers**

**complex → Special math numbers**

# **Example**

```python
students = 45      # int
temperature = 36.5 # float
z = 4 + 2j         # complex

print(students)
print(temperature)
print(z)
```

# **Real Life Example**
 **Number of books = int (whole number)**

 **Water bottle amount = float (decimal number)**

**Advanced science calculations = complex**
