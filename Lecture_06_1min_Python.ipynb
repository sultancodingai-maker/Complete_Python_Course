# **1️⃣ What are Data Types?**

**A data type defines the type of value a variable stores**


#**Think of it like different containers**

 **Integer (int) → Whole numbers**

 **Float (float) → Decimal numbers**

 **String (str) → Text inside quotes**

 **Boolean (bool) → True or False values**

# **Real Life Example**

**Different jars hold different things:**

**Candy jar → integers**

**Water bottle → float**

**Name tag → string**

**Light switch → boolean**

```python
age = 18 # int
price = 99.9 # float
name = "Ahmed" # string
is_online = True # boolean

print(age)
print(price)
print(name)
print(is_online)
```
