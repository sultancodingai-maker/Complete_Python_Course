# **Remaining Python Data Types**
**List (list)**

**Tuple (tuple)**

**Dictionary (dict)**

# **2️⃣ List (list)**

▶ **A list stores multiple**

**values in a single variable.**

 ▶ **Values are written**

 **inside square brackets [ ].**

# ▶**Real Life Example**

**Shopping list**

**Student names**

#▶**Important Rule**

**Lists use square brackets [ ].**

# **✅ Example**

```python
fruits = ["Apple", "Banana", "Mango"]

print(fruits)
```

# **3️⃣ Tuple (tuple)**

▶**A tuple stores multiple**

**values just like a list.**

▶**Values are written**

**inside parentheses ( ).**


#▶ **Real Life Example**

**Days of the weekend**

**Months of the year**

#▶**Important Rule**

**Tuples use parentheses ( ).**

# **✅ Example**

```python
colors = ("Red", "Blue", "Green")

print(colors)
```

# **4️⃣ Set (set)**

▶ **A set stores unique values.**

▶**Duplicate values are**


 **automatically removed.**


# ▶**Real Life Example**

**Unique student IDs**

**Unique email addresses**

# ▶**Important Rule**

**Sets do not allow duplicate values.**

# **✅ Example**

```python
numbers = {1, 2, 3, 3, 4}

print(numbers)
```

# **5️⃣ Dictionary (dict)**

▶**A dictionary stores data**

 **in key-value pairs.**

▶**Values are written inside**

 **curly brackets { }.**


#  ▶**Real Life Example**

**Student records**

**Employee information**

**Product details**

# ▶**Important Rule**

**Dictionaries store data as**

 **key : value pairs.**

# **✅ Example**

```python
student = {
    "name": "Ali",
    "age": 20
}

print(student)
```

# **Type, Use and Brackets**

| **Type** | **Use** | **Brackets** |
|----------|---------|--------------|
| **`List`** | Multiple values | **`[ ]`** |
| **`Tuple`** | Fixed values | **`( )`** |
| **`Set`** | Unique values | **`{ }`** |
| **`Dictionary`** | Key-value pairs | **`{ : }`** |
