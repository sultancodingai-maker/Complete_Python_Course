# **2️⃣ Keywords in Python**

▶ **Keywords are special reserved words in Python**.

▶ **These words already have fixed meanings.**

▶ **Python uses them to perform specific tasks**.


▶ **That’s why we cannot use them as:**

**Variable names**

**Function names**

**Class names**

# ▶  **Example**

**if, for, while, class, True, False**

# **Real World Example**
**Imagine a school**

**In school, some words already have fixed meanings:**

**“Principal” = Head of school**

**“Teacher” = Person who teaches**

**You cannot change their meanings.**

**For example:**
**You cannot call a student “Principal” because that word already has a special role.**
